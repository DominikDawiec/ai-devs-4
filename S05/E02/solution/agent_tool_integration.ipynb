# Agent for S05E02: Tool Integration

This notebook implements an agent that integrates various tools mentioned in the S05E02 lesson, including web browsing, file processing, and voice capabilities.

## Tools Integrated:
1. Web browsing with browser-use
2. Document processing with markitdown
3. Voice synthesis with elevenlabs
4. File system operations

```python
from browser_use import Agent
from firecrawl import FirecrawlApp
from elevenlabs import generate, play
from markitdown import MarkItDown
import os

class ToolIntegratedAgent:
    def __init__(self):
        # Initialize tools
        self.firecrawl = FirecrawlApp(api_key='your-firecrawl-key')
        self.markitdown = MarkItDown()
        # Note: ElevenLabs requires API key
        # self.elevenlabs_key = 'your-elevenlabs-key'
    
    def browse_web(self, url):
        """Browse a webpage using browser-use"""
        try:
            agent = Agent(
                task=f"Visit {url} and summarize the main content",
                llm=None  # Uses default
            )
            result = agent.run()
            return result
        except Exception as e:
            return f"Browsing error: {str(e)}"
    
    def scrape_web(self, url):
        """Scrape webpage content using Firecrawl"""
        try:
            scrape_result = self.firecrawl.scrape_url(url)
            return scrape_result['content']
        except Exception as e:
            return f"Scraping error: {str(e)}"
    
    def process_document(self, file_path):
        """Convert document to markdown using markitdown"""
        try:
            result = self.markitdown.convert(file_path)
            return result.text_content
        except Exception as e:
            return f"Document processing error: {str(e)}"
    
    def text_to_speech(self, text):
        """Convert text to speech using ElevenLabs"""
        try:
            audio = generate(
                text=text,
                voice="Bella",
                api_key=self.elevenlabs_key
            )
            play(audio)
            return "Audio played successfully"
        except Exception as e:
            return f"TTS error: {str(e)}"
    
    def list_directory(self, path):
        """List directory contents"""
        try:
            items = os.listdir(path)
            return items
        except Exception as e:
            return f"Directory listing error: {str(e)}"

agent = ToolIntegratedAgent()
```

```python
# Example: Web scraping
url = "https://example.com"
print("Scraping webpage:")
print(agent.scrape_web(url))
```

```python
# Example: Document processing
# Assuming we have a test file
test_file = "test.pdf"  # Replace with actual file
if os.path.exists(test_file):
    print("Processing document:")
    print(agent.process_document(test_file))
else:
    print("Test file not found")
```

```python
# Example: Directory listing
print("Listing current directory:")
print(agent.list_directory("."))
```
